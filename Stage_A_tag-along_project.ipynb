```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://github.com/HamoyeHQ/HDSC-Introduction-to-Python-for-machine-learning/files/7768140/FoodBalanceSheets_E_Africa_NOFLAG.csv?raw=true.csv', encoding= 'latin-1')
print(df)

agg=df.groupby('Item')[['Y2014','Y2017']].sum()  # groupby sum aggregation on Item
print(agg)
                                         
_mean=df['Y2015'].mean()   
print(round(_mean, 3))                            # mean and standard deviation across the whole dataset for
                                                 #the year 2015 to 3 decimal places?


standardDeviation=df['Y2015'].std()
print(round(standardDeviation, 3))

missingData= df['Y2016'].isnull()
TotalmissingData=missingData.count()
print(missingData.count)
ms=df['Y2016'].isna()
print(ms.count())

shape = (50,4460)

data = np.random.normal(size=shape)

data[:, 1000] += data[:, 2000]

df = pd.DataFrame(data)

c = df.corr().abs()

s = c.unstack()
so = s.sort_values(kind="quicksort")


uniq=df.groupby('Area').size()                    #the total number of unique countries in the dataset?
print(uniq.count())


ele=df.groupby('Element')[['Y2014','Y2015','Y2016','Y2017']].sum()  
print(ele)


ag=df.groupby('Element')[['Y2014']].sum()
print(ag)



elem=df.groupby('Element')[['Y2018']].sum()             #elements with the highest sum in 2018
print(elem)

Alg=df.groupby('Algeria')[['Y2014']].sum()
print(Alg)



```
